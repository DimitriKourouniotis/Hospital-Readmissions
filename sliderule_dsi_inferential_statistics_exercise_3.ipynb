# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
#Create the abbreviated dataset
df_shortened = clean_hospital_read_df[81:-3]
```

Calculate the Pearson Correlation Coefficient

```python
pearson_r = df_shortened['Number of Discharges'].corr(df_shortened['Excess Readmission Ratio'])
print('Pearson R coefficient: {} - only a marginal correlation between Discharge and Readmission'.format(pearson_r))
```

<B><i> H<sub>0</sub></B></i> Null Hypothesis: The average number of Excess Rate of Readmissions from Hospitals < 100 Discharges is NOT equal to the Excess Rate of Readmissions from Hospitals > 1000 Discharges.
 
<B><i> H<sub>1</sub></B></i> Alternative Hypothesis: The average number of Excess Rate of Readmissions from Hospitals < 100 Discharges is equal to the Excess Rate of Readmissions from Hospitals > 1000 Discharges.

****

Filter from the cleaned data set two groups, small and large hospitals.

```python
#create 2 new datasets to compare, small and large hospitals
small_df = df_shortened[df_shortened['Number of Discharges'] <100 ]
large_df = df_shortened[df_shortened['Number of Discharges'] >1000 ]
```

```python
from scipy import stats
from scipy.stats.distributions import norm
```

```python
#calculate the different mean Excess Readmission ratio
mean_small = small_df['Excess Readmission Ratio'].mean()
mean_large = large_df['Excess Readmission Ratio'].mean()
print('Small Hospital Mean Excess Readmission Ratio: {}'.format(mean_small))
print('Large Hospital Mean Excess Readmission Ratio: {}'.format(mean_large))
```

```python
md = mean_small - mean_large
print('Mean difference: {}'.format(md))
```

```python
X = small_df['Excess Readmission Ratio']
Y = large_df['Excess Readmission Ratio']
```

```python
#Standard Error
se = np.sqrt(np.var(X)/len(X) + np.var(Y)/len(Y))
print('Standard Error: {} '.format(se))
```

```python
#Lower and Upper confidence limits
lcl,ucl = md-2*se,md+2*se
lcl,ucl
```

```python
#Z-score
z=md/se
print('z-score: {}'.format(z))
```

```python
#P-value
pvalue = -2*norm.cdf(-np.abs(z))
print('p-value: {}'.format(pvalue))
```

Margin of Error : z-value(99%) x standard error:

z-value for 99% confidence is 2.58 <br>
standard error has been calculated above as 0.005807746245524579

```python
print('Marigin of Error: {}'.format(2.58*se))
```

With the p-value from the test being less than <0.01 we can REJECT the null hypothesis

From a statistical point of view there is not enough difference, but the plot does show a wide variety of variance at the lower levels and that should be looked into further.
